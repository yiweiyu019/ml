```python
def data_split(sp,path,flag,class1):
    final_data = {"class":[]}
    final_data["class"].append(class1)
    for j in range(sp):
        for i in range(1,6+1):
            final_data['max'+str(6*j+i)] = []
            final_data['mean'+str(6*j+i)] = []
            final_data['third_q'+str(6*j+i)] = []
        
    if flag == 0:   
        data_b = pd.read_csv(path,skiprows=4)
    else:
        data_b = pd.read_csv(path,sep ='\s+',skiprows=5,names = ["Columns: time","avg_rss12","var_rss12","avg_rss13","var_rss13","avg_rss23","var_rss23"])
    if(data_b.shape[0]%sp == 0):
        part_size = data_b.shape[0]//sp
        total_part = data_b.shape[0] // part_size
        for i in range(0,total_part):
            chunck = data_b.iloc[i*part_size:i*part_size+part_size]
            final_data['mean'+str(6*i+1)].append(chunck.avg_rss12.describe().loc["mean"])
            final_data['max'+str(6*i+1)].append(chunck.avg_rss12.describe().loc["max"])
            final_data['third_q'+str(6*i+1)].append(chunck.avg_rss12.describe().loc["75%"])
            
            final_data['mean'+str(6*i+2)].append(chunck.var_rss12.describe().loc["mean"])
            final_data['max'+str(6*i+2)].append(chunck.var_rss12.describe().loc["max"])
            final_data['third_q'+str(6*i+2)].append(chunck.var_rss12.describe().loc["75%"])
            
            final_data['mean'+str(6*i+3)].append(chunck.avg_rss13.describe().loc["mean"])
            final_data['max'+str(6*i+3)].append(chunck.avg_rss13.describe().loc["max"])
            final_data['third_q'+str(6*i+3)].append(chunck.avg_rss13.describe().loc["75%"])
            
            final_data['mean'+str(6*i+4)].append(chunck.var_rss13.describe().loc["mean"])
            final_data['max'+str(6*i+4)].append(chunck.var_rss13.describe().loc["max"])
            final_data['third_q'+str(6*i+4)].append(chunck.var_rss13.describe().loc["75%"])
            
            final_data['mean'+str(6*i+5)].append(chunck.avg_rss23.describe().loc["mean"])
            final_data['max'+str(6*i+5)].append(chunck.avg_rss23.describe().loc["max"])
            final_data['third_q'+str(6*i+5)].append(chunck.avg_rss23.describe().loc["75%"])
            
            final_data['mean'+str(6*i+6)].append(chunck.var_rss23.describe().loc["mean"])
            final_data['max'+str(6*i+6)].append(chunck.var_rss23.describe().loc["max"])
            final_data['third_q'+str(6*i+6)].append(chunck.var_rss23.describe().loc["75%"])
            
            
    else:  
        part_size = int(np.ceil(data_b.shape[0]/sp))
        total_part = data_b.shape[0] // part_size
        for i in range(0,total_part):
            chunck = data_b.iloc[i*part_size:i*part_size+part_size]
            final_data['mean'+str(6*i+1)].append(chunck.avg_rss12.describe().loc["mean"])
            final_data['max'+str(6*i+1)].append(chunck.avg_rss12.describe().loc["max"])
            final_data['third_q'+str(6*i+1)].append(chunck.avg_rss12.describe().loc["75%"])
            
            final_data['mean'+str(6*i+2)].append(chunck.var_rss12.describe().loc["mean"])
            final_data['max'+str(6*i+2)].append(chunck.var_rss12.describe().loc["max"])
            final_data['third_q'+str(6*i+2)].append(chunck.var_rss12.describe().loc["75%"])
            
            final_data['mean'+str(6*i+3)].append(chunck.avg_rss13.describe().loc["mean"])
            final_data['max'+str(6*i+3)].append(chunck.avg_rss13.describe().loc["max"])
            final_data['third_q'+str(6*i+3)].append(chunck.avg_rss13.describe().loc["75%"])
            
            final_data['mean'+str(6*i+4)].append(chunck.var_rss13.describe().loc["mean"])
            final_data['max'+str(6*i+4)].append(chunck.var_rss13.describe().loc["max"])
            final_data['third_q'+str(6*i+4)].append(chunck.var_rss13.describe().loc["75%"])
            
            final_data['mean'+str(6*i+5)].append(chunck.avg_rss23.describe().loc["mean"])
            final_data['max'+str(6*i+5)].append(chunck.avg_rss23.describe().loc["max"])
            final_data['third_q'+str(6*i+5)].append(chunck.avg_rss23.describe().loc["75%"])
            
            final_data['mean'+str(6*i+6)].append(chunck.var_rss23.describe().loc["mean"])
            final_data['max'+str(6*i+6)].append(chunck.var_rss23.describe().loc["max"])
            final_data['third_q'+str(6*i+6)].append(chunck.var_rss23.describe().loc["75%"])


       ####
        chunck = data_b.iloc[total_part*part_size:]
        final_data['mean'+str(6*total_part+1)].append(chunck.avg_rss12.describe().loc["mean"])
        final_data['max'+str(6*total_part+1)].append(chunck.avg_rss12.describe().loc["max"])
        final_data['third_q'+str(6*total_part+1)].append(chunck.avg_rss12.describe().loc["75%"])

        final_data['mean'+str(6*total_part+2)].append(chunck.var_rss12.describe().loc["mean"])
        final_data['max'+str(6*total_part+2)].append(chunck.var_rss12.describe().loc["max"])
        final_data['third_q'+str(6*total_part+2)].append(chunck.var_rss12.describe().loc["75%"])

        final_data['mean'+str(6*total_part+3)].append(chunck.avg_rss13.describe().loc["mean"])
        final_data['max'+str(6*total_part+3)].append(chunck.avg_rss13.describe().loc["max"])
        final_data['third_q'+str(6*total_part+3)].append(chunck.avg_rss13.describe().loc["75%"])

        final_data['mean'+str(6*total_part+4)].append(chunck.var_rss13.describe().loc["mean"])
        final_data['max'+str(6*total_part+4)].append(chunck.var_rss13.describe().loc["max"])
        final_data['third_q'+str(6*total_part+4)].append(chunck.var_rss13.describe().loc["75%"])

        final_data['mean'+str(6*total_part+5)].append(chunck.avg_rss23.describe().loc["mean"])
        final_data['max'+str(6*total_part+5)].append(chunck.avg_rss23.describe().loc["max"])
        final_data['third_q'+str(6*total_part+5)].append(chunck.avg_rss23.describe().loc["75%"])

        final_data['mean'+str(6*total_part+6)].append(chunck.var_rss23.describe().loc["mean"])
        final_data['max'+str(6*total_part+6)].append(chunck.var_rss23.describe().loc["max"])
        final_data['third_q'+str(6*total_part+6)].append(chunck.var_rss23.describe().loc["75%"])
        
    return pd.DataFrame(final_data)
        
```

```python
def data_loading(n_pieces):
    train_seg = pd.DataFrame()
    test_seg = pd.DataFrame()


    for i in ["standing","cycling","walking","sitting","lying"]:
        path = "../Data/AReM/" +i
        for ff in sorted(os.listdir(path),key = lambda x:int(x.split(".csv")[0].split("dataset")[1])):
            #print("../Data/AReM/"+i+'/'+ff)
            if(int(ff.split(".csv")[0].split("dataset")[1])<=3):            
                test_seg = test_seg.append(data_split(n_pieces,"../Data/AReM/"+i+'/'+ff,flag = 0,class1 = i),ignore_index = True)

            else:
                train_seg = train_seg.append(data_split(n_pieces,"../Data/AReM/"+i+'/'+ff,flag = 0,class1 = i),ignore_index = True)


    for i in ["bending1","bending2"]:
        path = "../Data/AReM/" +i
        for ff in sorted(os.listdir(path),key = lambda x:int(x.split(".csv")[0].split("dataset")[1])):
            #print("../Data/AReM/"+i+'/'+ff)
            if(int(ff.split(".csv")[0].split("dataset")[1])<=2):
                test_seg = test_seg.append(data_split(n_pieces,"../Data/AReM/"+i+'/'+ff,flag = 0,class1 = i),ignore_index = True)
            else:
                if (int(ff.split(".csv")[0].split("dataset")[1])) == 4 and i == "bending2":    
                    train_seg = train_seg.append(data_split(n_pieces,"../Data/AReM/"+i+'/'+ff,flag = 1,class1 = i),ignore_index = True)
                else:
                    train_seg = train_seg.append(data_split(n_pieces,"../Data/AReM/"+i+'/'+ff,flag = 0,class1 = i),ignore_index = True)
    return train_seg,test_seg
```

```python
import matplotlib.pyplot as plt
from sklearn import svm, datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc
from sklearn.multiclass import OneVsRestClassifier
from itertools import cycle
plt.style.use('ggplot')

iris = datasets.load_iris()
X = iris.data
y = iris.target

# Binarize the output
y = label_binarize(y, classes=[0, 1, 2])
n_classes = y.shape[1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.5, random_state=0)

classifier = OneVsRestClassifier(svm.SVC(kernel='linear', probability=True,
                                 random_state=0))
y_score = classifier.fit(X_train, y_train).decision_function(X_test)

fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
colors = cycle(['blue', 'red', 'green'])
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=1.5,
             label='ROC curve of class {0} (area = {1:0.2f})'
             ''.format(i, roc_auc[i]))
plt.plot([0, 1], [0, 1], 'k--', lw=1.5)
plt.xlim([-0.05, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic for multi-class data')
plt.legend(loc="lower right")
plt.show()
```
