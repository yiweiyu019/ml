```python
x = sm.add_constant(data_core['AT'].values)
model1 = sm.OLS(data_core['PE'].values,x).fit()
influ = model1.get_influence()
temp1 = influ.cooks_distance[0]
test_id = []
ind = -1
for i in temp1:
    ind+=1
    if i > 4/9568:
        test_id.append(ind)
        
####
outliers_x = []
outliers_y = []
for i in test_id:
    outliers_x.append(data_core['AT'][i])
    outliers_y.append(data_core['PE'][i])
    
###
linear_model_at = LinearRegression()
linear_model_at.fit(data_core['AT'].values.reshape(-1,1),data_core['PE'].values.reshape(-1,1))
plt.rcParams["figure.figsize"] = (18,9)
plt.scatter(data_core['AT'].values.reshape(-1,1),data_core['PE'].values.reshape(-1,1))
plt.scatter(outliers_x,outliers_y,color='r')
plt.plot(data_core['AT'].values.reshape(-1,1),linear_model_at.predict(data_core['AT'].values.reshape(-1,1)))
plt.show()
```

```python
out1 = pd.DataFrame({'AT':outliers_x,'PE':outliers_y})
print('Outliers detected from model between AT and PE')
out1
```

<h5>1. For each of parts (a) through (d), indicate whether we would generally expect the performance of a flexible statistical learning method to be better or worse than an inflexible method. Justify your answer.</h5>

<p>
    <b>(a)</b> The sample size n is extremely large, and the number of predictors p is small.</p>

<p>Flexible statistical learning method is better. The inflexible statistical learning method will cause underfitting on the extremely large sample size with a small p,but the flexible method will be able to extract more meaningful infomation based on the large n</p>

<p><b>(b)</b> The number of predictors p is extremely large, and the number of observations n is small.</p>

<p>Flexible statistical learning method is worse. When the number of obersvations n is small, the flexible method could quite possiblely be overfitting, and it'll fit too much on the noise. </p>


<p><b>(c)</b> The relationship between the predictors and response is highly non-linear.</p>

<p>Flexible statistical learning method is better. The flexible method will be able to find non-linear pattern among the non-linear relationship between the predictors and response. </p>


<p><b>(d)</b>The variance of the error terms, i.e. Ïƒ2 = Var(Îµ), is extremely high.</p>

<p>Flexible statistical learning method is worse. The flexible method will fit in toom uch noise(error terms) because of the high variance, and the inflexible method will capture less noise in that case. </p>
